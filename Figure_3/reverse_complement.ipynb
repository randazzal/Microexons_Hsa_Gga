```python
from Bio import SeqIO
```

```python
def reverse_complement_with_gaps(seq):
    complement = str.maketrans("ACGTacgt", "TGCAtgca")
    # Reverse the sequence but keep gaps ('-') in the same position
    rev_comp = "".join(seq.translate(complement)[::-1])
    return rev_comp
```

```python
input_file = "/data2/lackey_lab/DownloadedSequenceData/randazza/APBB2_alignment.fasta"
output_file = "APBB2_reverse_complement.fasta"
```

```python
with open(output_file, "w") as out_f:
    for record in SeqIO.parse(input_file, "fasta"):
        record.seq = reverse_complement_with_gaps(str(record.seq))
        SeqIO.write(record, out_f, "fasta")
```
